```
from sympy import *
b=symbols('b')
f=b*4+3*b*2+10
z=diff(f,b) 
print(z)
B=int(input("enter b value:"))
itr=1 
n=0.1 
epoch=2
while (itr <= epoch) :
   Z=z.subs(b,B)
   print("z",Z)
   deltax=-n*Z
   print("delta(b) =",deltax)
   newx = B + deltax
   print("b =",newx)
   itr= itr + 1
   B=newx
print("final x =",B)
print("Minimum value is",f.subs(b,B))
```
